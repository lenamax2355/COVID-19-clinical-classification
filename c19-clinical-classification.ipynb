# Objectif

Le dataset contient +5000 cas cliniques provenant de l'hôpital Israelita Albert Einstein (São Paulo, Brésil).

- **Objectif:** Prédire si une personne est infectée en fonction des données cliniques disponibles
- **Métrique:** Classes sont déséquilibrées, donc:
    - $Précision = \frac{TP}{TP + FP}$, permet de réduire au maximum le taux de Faux Positifs (FP).
    - $Recall (sensibilité) = \frac{TP}{TP + FN}$, permet de réduire au maximum le taux de Faux Négatifs (FN).
    - **Score F1**: fait le rapport entre précision et recall.

La métrique est à définir, mais à titre d'exemple ici on peut choisir: F1 > 50% et Recall > 70%

#### Encodage

Génération de la liste des colonnes après encodage

```python
def make_encoded_column_list(column_list):
    """
    A partir de la liste de colonnes originale, retourne la liste de colonnes encodées
    exemple:
    - liste originale:  ['x0', 'x1'] 
    - liste encodée: ['x0_detected', 'x0_not_detected', 'x0_unk', 'x1_detected', 'x1_not_detected', 'x1_unk']
    
    A utiliser pour cat_features_2_state
    """
    return [i + j for i in column_list for j in ["_detected", "_not_detected", "_unk"]]

processed_column_list = (y_column
               + int_features
               + cat_features_1_state
               + make_encoded_column_list(cat_features_2_state))
```

Encodage, récupréation des résultats dans un dataframe avec noms de colonnes

```python
cat_1_states_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="unk"),
                                      OrdinalEncoder())

cat_2_states_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="unk"),
                                      OneHotEncoder())

preprocessor = make_column_transformer((OrdinalEncoder(), y_column),
                                       (OrdinalEncoder(), int_features),
                                       (cat_1_states_pipeline, cat_features_1_state),
                                       (cat_2_states_pipeline, cat_features_2_state))

transformed_data = preprocessor.fit_transform(dataset_no_empty)
dataset_preprocessed = pd.DataFrame(transformed_data, columns=processed_column_list)
```

```python
def make_encoded_column_list(column_list):
    """
    A partir de la liste de colonnes originale, retourne la liste de colonnes encodées
    exemple:
    - liste originale:  ['x0', 'x1'] 
    - liste encodée: ['x0_not_detected', 'x0_unk', 'x1_not_detected', 'x1_unk']
    
    A utiliser pour cat_features_2_state
    """
    return [i + j for i in column_list for j in ["_not_detected", "_unk"]]

processed_column_list = (y_column
               + int_features
               + cat_features_1_state
               + make_encoded_column_list(cat_features_2_state))

cat_1_states_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="unk"),
                                      OrdinalEncoder())

cat_2_states_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="unk"),
                                      OneHotEncoder(drop="first"))

preprocessor = make_column_transformer((OrdinalEncoder(), y_column),
                                       (OrdinalEncoder(), int_features),
                                       (cat_1_states_pipeline, cat_features_1_state),
                                       (cat_2_states_pipeline, cat_features_2_state))

transformed_data = preprocessor.fit_transform(dataset_no_empty)
dataset_preprocessed = pd.DataFrame(transformed_data, columns=processed_column_list)
```

```python
dataset = pd.read_excel("fichiers/dataset_covid19.xlsx").drop("Patient ID", axis=1)

# dict dtype:colone
columns_types = dataset.columns.to_series().groupby(dataset.dtypes).groups

# on convertit toutes les colonnes catégoriques en str (car présence de nombres aussi)
categorical_colomns = columns_types[np.dtype('O')]
dataset[categorical_colomns] = dataset[categorical_colomns].astype(str)
```

```python
X = dataset.drop("SARS-Cov-2 exam result", axis=1, inplace=False)
y = dataset["SARS-Cov-2 exam result"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV

int_features = make_column_selector(dtype_include="number")#types_colonnes[np.dtype('int64')]
float_features = make_column_selector(dtype_include="number")#types_colonnes[np.dtype('float64')]
categorical_features = make_column_selector(dtype_include="object")#types_colonnes[np.dtype('O')]

int_pipeline = make_pipeline(SimpleImputer(strategy="mean"), StandardScaler(), PCA())
float_pipeline = make_pipeline(SimpleImputer(strategy="mean"), StandardScaler(), PCA())
categorical_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="unk"), OneHotEncoder())

preprocessor = make_column_transformer((categorical_pipeline, categorical_features),
                                       (int_pipeline, int_features),
                                       (float_pipeline, float_features))

model = make_pipeline(preprocessor, SVC())

model
```

```python
params = {
    "svc__kernel": ["poly", "rbf"]
}

grid = GridSearchCV(model, param_grid=params, cv=2, verbose=True, n_jobs=-1)
grid.fit(X_train, y_train)
```
